```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/it.csv')
df.head(10)
```

### Определите среднюю з.п. по данным в рублях

```python
dfRub = df[df['З/п в валюте найма'].str.contains('₽')]
meanWage = dfRub['З/п в валюте найма'].str.translate(str.maketrans({'₽': '', ',': '.', u'\xa0': ''})).astype(float).mean()
print('Средняя з.п.:',str(meanWage) + '₽')
```

### Определите технологию с максимальной з.п. по данным в рублях

```python
dfRub.groupby(['Технология'])['З/п в валюте найма'].max()
# Ответ: QA -> 375 000,00 ₽ (Но база странная, так как насколько я понимаю QA это не технология, а профессия) Если брать технологии, то это JS или Java, что не удивительно
```

### Определите возраст программистов с максимальной з.п. по данным в рублях

```python
dfRub.groupby(['Дата рождения'])['З/п в валюте найма'].max()
# Ответ: 34 (16.03.1986 -> 375 000,00 ₽)
```
