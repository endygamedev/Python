```python
import pandas as pd
import json

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/dm-fedorov/pandas_basic/master/data/tinkoff_hh.json"
```

```python
with open('data.txt', encoding='utf8') as data:
    file = data.readlines()
```

### Делаем DataFrame 

```python
file
```

```python
lst = [json.loads(element) for element in file if 'index' not in element]
lst
```

```python
df = pd.DataFrame(lst)
df.head()
```

### Распеределение зарплат №1

```python
dfVacancyIncome = df[["vacancy","income","region"]][df["income"] != "не указан"].drop_duplicates()
dfVacancyIncome
```

```python
dfVacancyIncome['vacancy'] = dfVacancyIncome['vacancy'].str.lower()
dfVacancyIncome = dfVacancyIncome.drop_duplicates()
dfVacancyIncome
```

```python
dfVacancyIncome["vacancy"].value_counts()
```

```python
dfVacancyIncome[dfVacancyIncome["vacancy"] == "представитель тинькофф"]
```

### Распределение зарплат №2 (в тупую)

```python
df = pd.DataFrame(lst)
dfCopy = df
```

```python
def income_to_number(row):
    s = row['income'].split()
    if row['income'] == 'не указан':
        return 0
    elif 'до' in row['income']:
        return int(s[1])*1000
    else:
        return int(s[1])*1000
```

```python
dfCopy['income'] = dfCopy.apply(income_to_number, axis=1)
dfCopy
```

```python
seriesVacancy = dfCopy.groupby(['vacancy'])['income'].mean().sort_values(ascending=False)
seriesVacancy
```

```python
plt.title('Профессии в Tinkoff, которые в среднем зарабатывают больше 100 000 руб.')

seriesVacancy[seriesVacancy>100000].plot.barh()

plt.ylabel('Должность')
plt.xlabel('Зарплата')
ax = plt.gca()
ax.tick_params(axis='x', colors='blue')
ax.tick_params(axis='y', colors='green')
plt.show()
```

Определите статистику встречаемости отдельных слов в поисковых фразах. Это позволит понять тематику данного сайта и настроить показ рекламы.

За помощью обратитетесь к [инструкции](https://github.com/dm-fedorov/pandas_basic/blob/master/%D0%B1%D1%8B%D1%81%D1%82%D1%80%D0%BE%D0%B5%20%D0%B2%D0%B2%D0%B5%D0%B4%D0%B5%D0%BD%D0%B8%D0%B5%20%D0%B2%20pandas/10.%20%D0%9A%D0%B0%D0%BA%20%D0%BC%D0%B0%D0%BD%D0%B8%D0%BF%D1%83%D0%BB%D0%B8%D1%80%D0%BE%D0%B2%D0%B0%D1%82%D1%8C%20%D1%82%D0%B5%D0%BA%D1%81%D1%82%D0%BE%D0%B2%D1%8B%D0%BC%D0%B8%20%D0%B4%D0%B0%D0%BD%D0%BD%D1%8B%D0%BC%D0%B8%3F.ipynb) и возможностям модуля [pymorphy2](https://github.com/dm-fedorov/python_basic/blob/master/pymorphy2.ipynb).

```python
lstPhrases = [j for i, j in enumerate(df["Поисковая фраза"].str.lower())]
lstPhrases
```

```python
splitedPhrases = list(map(lambda x: x.split(), lstPhrases))
```

```python
flat_list = [morph.parse(item)[0].normal_form for sublist in splitedPhrases for item in sublist]
flat_list
```

```python
seriesPhrases = pd.Series(flat_list)
```

```python
plt.title('Слова, которые встречаются больше 2000 раз')

seriesPhrases.value_counts()[seriesPhrases.value_counts()>2000].plot.bar()

plt.ylabel('Кол-во встречаемости слова')
plt.xlabel('Слова')
ax = plt.gca()
ax.tick_params(axis='x', colors='blue')
ax.tick_params(axis='y', colors='green')
plt.show()
```
