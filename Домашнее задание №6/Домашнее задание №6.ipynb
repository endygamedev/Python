### Упраженение 11.1
**Создайте тип данных Human, определите атрибуты и методы для работы с ним.**

```python
class Human:
    def __init__(self, name = 'Без имени'):
        self.__name = name
        print(f'Родился человек с именем {self.__name}')

    def eating(self):
        print(f'{self.__name} ест')    

    def sleeping(self):
        print(f'{self.__name} спит') 

    def working(self):
        print(f'{self.__name} работает') 
        
    def relaxing(self):
        print(f'{self.__name} отдыхает') 

    def growing(self):
        print(f'{self.__name} отмечает день рождения')
        
class Life:
    def life(self, human, years=70):
        while years:            
            human.growing()
            human.eating()
            human.sleeping()
            human.working()
            human.relaxing()
            years -= 1
```

```python
petr = Human('Петр')
life_petr = Life()
life_petr.life(petr)
```

### Упраженение 11.2 (Game Architecture)

**Провести объектно-ориентированный анализ задачи и определить объекты (создать классы с заданным поведением по аналогии с КОШКОЙ), которые потребуется использовать:**

*- создается компьютерная игра, поддерживающая одновременное управление несколькими персонажами;*

*- каждый персонаж умеет бегать, стрелять и собирать предметы;*

*- некоторые персонажи умеют летать (использовать наследование классов);*

*- каждый тип действий сопровождается определенной анимацией (подсказка: изменением координат);*

*- собираемые предметы могут менять характеристики действий и соответствующую анимацию (например, вдвое ускорять или, наоборот, замедлять бег персонажа).*

```python
class Character:
    # Атрибуты класса
    count_char = 0 # Считаем кол-во персонажей, которыми мы будем управлять
    animation = ["AnimationStay","AnimationRun","AnimationShoot", "AnimationGetItem", "AnimationFlying"] # Возможные "анимации" персонажа
    
    def __init__(self, name = 'Персонаж без имени', x = 0, y = 0, power = 10, speed = 5, items = 1):
        '''
        Конструктор игрового персонажа
        '''
        self.__name = name # Имя персонажа
        self._x = x # Координаты по x
        self._y = y # Координаты по y
        self._power = power # Сила
        self._speed = speed # Скорость
        self._items = items # Вещи
        Character.count_char += 1
        print(f'Создался новый персонаж {self.__name}: {Character.animation[0]}')
    
    def run(self, delta_x = 1, delta_y = 1):
        '''
        Метод бега персонажа
        '''
        self._x += delta_x
        self._y += delta_y
        print(f'Персонаж {self.__name} бежит, его координаты {self._x, self._y}: {Character.animation[1]}')
    
    def shoot(self):
        '''
        Метод стрельбы персонажа
        '''
        print(f'Пиу-Пиу, персонаж {self.__name} выстрелил: {Character.animation[2]}')
    
    def get_items(self, item):
        '''
        Метод собирания предметов
        '''
        if item == 'SuperSpeedSalve': # Магическая пробирка для повышении скорости персонажа
            self._speed *= 2
        elif item == 'SuperPowerSalve': # Магическая пробирка для повышении силы персонажа
            self._power *= 1.5
        else:
            self._items += 1 # Если вешь обычная, то ложим в рюкзак персонажа
        print(f'Персонаж {self.__name} подобрал предмет {item}: {Character.animation[3]}')
        
class FlyingCharacter(Character):
    def __init__(self, name = 'Персонаж без имени', x = 0, y = 0, power = 20, speed = 10, items = 1):
        '''
        Конструктор летающего персонажа
        '''
        Character.__init__(self, name = name, x = x, y = y, power = power, speed = speed, items = items)
        
```
